# Task 1 – Conceptual Understanding

## Question 01. What is the difference between "Love" and "love" in NLP?

In Natural Language Processing, **"Love" and "love" are treated as different words** because computers are case-sensitive. This means uppercase and lowercase words are considered separate tokens.

For example:

* "Love" may appear at the beginning of a sentence.
* "love" may appear in the middle of a sentence.

If we do not convert text to lowercase, the model will treat them as two different words, which increases vocabulary size and reduces model efficiency.
Therefore, **lowercasing is an important preprocessing step in NLP** to maintain consistency.


## Question 02. What happens if stopwords are not removed?

Stopwords are common words like **is, the, are, in, on, at, a, an** which do not carry significant meaning.

If stopwords are not removed:

* The dataset will contain many unnecessary words.
* Processing time will increase.
* Memory usage will increase.
* Machine learning models may focus on unimportant words.
* Accuracy of text analysis may decrease.
* Important keywords may get less importance.

So, removing stopwords helps in **reducing noise and improving model performance**.

## Question 03. Provide two real-world scenarios where removing stopwords can be harmful.

Removing stopwords is not always good. Sometimes stopwords carry important meaning.

### Scenario 1 – Sentiment Analysis

Example sentences:

* "I **do not** like this product."
* "I like this product."

If we remove stopwords like **not**, both sentences become:

* "I like this product"

This changes the meaning completely and gives wrong sentiment.

### Scenario 2 – Question Answering / Search Engines

Example:

* "What is the capital of India?"
  If stopwords like **what, is, the, of** are removed, the sentence becomes:
* "capital India"

This loses the actual question structure and meaning.

So in tasks like **sentiment analysis, chatbots, search engines**, removing stopwords can be harmful.

## Question 04. Explain the difference between stemming and lemmatization.


Stemming and lemmatization are both text preprocessing techniques used in Natural Language Processing to reduce words to their base or root form, but they work in different ways. Stemming is a simple and fast method that removes suffixes from words by chopping off word endings without considering the meaning of the word. Because of this, stemming sometimes produces root words that are not actual dictionary words. For example, words like running, played, studies may be reduced to run, play, studi using stemming. On the other hand, lemmatization is a more advanced and accurate method that reduces a word to its base form (lemma) using vocabulary and morphological analysis, meaning it considers the context and converts the word into a meaningful dictionary form. For example, running, better, studies are converted to run, good, study using lemmatization. In summary, stemming is faster but less accurate, while lemmatization is slower but produces meaningful and correct root words.

# Task 2 – Advanced Preprocessing Function

````python
import re

def preprocess_text(text):
    
    # 1. Convert to lowercase
    text = text.lower()
    
    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # 3. Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)
    
    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 5. Handle repeated characters (soooo → soo, goooood → good)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # 6. Remove special characters / punctuation
    text = re.sub(r'[^a-z\s]', '', text)
    
    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 8. Remove very short tokens (length ≤ 2), keep 'no', 'not'
    tokens = text.split()
    filtered_tokens = []
    
    for word in tokens:
        if len(word) > 2 or word in ['no', 'not']:
            filtered_tokens.append(word)
    
    # Join back
    text = " ".join(filtered_tokens)
    
    return text
````

````python
texts = [
    "I have 2 dogs",
    "This    is     good",
    "soooo goooood!!!",
    "WOW!!! This is GREAT!!!",
    "Visit http://example.com now",
    "Contact me at abc@gmail.com",
    "I am not happy"
]

for t in texts:
    print("Original :", t)
    print("Processed:", preprocess_text(t))
    print()
````

# Task 4 – Token Analytics

````python
def token_analytics(tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    
    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0
    
    return total_tokens, unique_tokens, avg_length
````

````python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for s in sentences:
    tokens, clean_sent = preprocess_text(s)
    total, unique, avg_len = token_analytics(tokens)
    
    print("Original Text        :", s)
    print("Cleaned Sentence     :", clean_sent)
    print("Tokens               :", tokens)
    print("Total Tokens         :", total)
    print("Unique Tokens        :", unique)
    print("Average Token Length :", round(avg_len,2))
    print("-"*60)
````

# Task 6 – Full Pipeline Function

def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)
    
    result = {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
    
    return result

sentences = [

]

output = full_pipeline(sentences)

print("All Tokens:")

A full NLP pipeline is a system that takes raw text input and processes it through multiple preprocessing steps to produce clean text and tokens that can be used for machine learning models.

Instead of running preprocessing manually for each sentence, we build a pipeline function that automatically processes a list of sentences and returns structured output.

# Task 7 – Error Handling

````python
import re

def preprocess_text(text):
    
    # Error Handling: Empty or None
    if text is None:
        return [], ""
    
    if not isinstance(text, str):
        text = str(text)
    
    if text.strip() == "":
        return [], ""
    
    # Lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # Remove special characters (including emojis)
    text = re.sub(r'[^a-z\s]', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Tokenization
    tokens = text.split()
    
    # Remove short tokens except 'no', 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
    
    cleaned_sentence = " ".join(tokens)
    
    return tokens, cleaned_sentence
````

````python
test_cases = [
    "",
    "😍😍🔥🔥",
    "123456789",
    "   ",
    None
]

for t in test_cases:
    tokens, clean_sent = preprocess_text(t)
    
    print("Input:", t)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean_sent)
    print("-"*40)
````

Error handling was implemented in the preprocessing function to ensure the system works correctly even when invalid or unusual input is provided. The function checks for empty strings, None values, strings containing only emojis, and strings containing only numbers. In such cases, the function returns empty tokens and an empty cleaned sentence instead of causing errors. This makes the preprocessing pipeline robust and suitable for real-world NLP applications where input data may be noisy or incomplete.
